```python
from pydub import AudioSegment

#importing file from location by giving its path
inputlist=[]
for i in range(no_of_videos):
  if i==0:
    suffix=""
  else:
    suffix="("+str(i)+")"
  inputlist.append("/content/audio"+suffix+".mp3")

outputlist=[]
for i in range(no_of_videos):
  if i==0:
    suffix=""
  else:
    suffix="("+str(i)+")"
  outputlist.append("/content/portion"+suffix+".mp3")

for i in range(no_of_videos):
  sound = AudioSegment.from_mp3(inputlist[i])

  #Selecting Portion we want to cut
  StrtMin = 0
  StrtSec = 0

  EndMin = 0
  EndSec = trim_length

  # Time to milliseconds conversion
  StrtTime = StrtMin*60*1000+StrtSec*1000
  EndTime = StrtMin*60*1000+EndSec*1000

  # Opening file and extracting portion of it
  extract = sound[StrtTime:EndTime]

  # Saving file in required location
  extract.export(outputlist[i], format="mp3")

  # new file portion.mp3 is saved at required location
```

```python

# enter the path of your audio file
sound1 = AudioSegment.from_mp3("/content/portion.mp3")


# merge two audio

sound3 = sound1

outputlist=[]
for i in range(no_of_videos):
  suffix="("+str(i+1)+")"
  outputlist.append("/content/portion"+suffix+".mp3")

for i in range(no_of_videos-2):
  sound2 = AudioSegment.from_mp3(outputlist[i])
  sound3 = sound3.append(sound2,crossfade=1000)


sound3.export("/content/mashupyay.mp3",format="mp3")

```
